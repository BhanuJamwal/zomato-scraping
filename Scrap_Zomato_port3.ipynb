```python
def fwrites(dictionary,res_id):
    import codecs
    open=codecs.open
    #f=open("menu.json",'a')
    l=len(dictionary[res_id])
    
    s= '{"id" : "%s", "items: ['%(res_id)
    for i in dictionary[res_id]:
       s+='{"name": "%s"},'%(i)
    s+="]},\n"
    with open('menu.json', 'a', encoding='utf-8') as f:
            f.write(s)
   
    f.close()
def jsonwrites(a_dict): 
    #print(a_dict)
    import codecs
    open=codecs.open
    with open("menu.json", 'a',encoding='utf-8') as f:
        json_obj = json.dump(a_dict,f)
        f.write(",\n")
        f.close()

def pagelink(links,done_rest):
    #print("In PAGE LINK")
    browser.get(links)
    browser.implicitly_wait(40)
    html=browser.page_source
    soop=BeautifulSoup(html,'html.parser')
    path=soop.find_all('a',{'class':'res_name'})
    for link in path:
        restlink(link.get('href'),done_rest)    

def restlink(links,done_rest):
    #print("in rest link")
    browser.get(links)
    browser.implicitly_wait(60)
    html_last=browser.page_source
    soop=BeautifulSoup(html_last,'html.parser')
    path=soop.find_all('a',{'class':'item respageMenu-item o2_link restaurant-tab-item-jumbo-track'})
    #if path:
     #   pass;
    #else: print("no path")
    for link in path:
        menulink(link.get('href'),done_rest)
        
def menulink(links,done_rest):
    #print("\nIn MENU LINK\n")
    print(len(done_rest))
    menu=dict()
    browser.get(links)
    browser.implicitly_wait(60)
    html_final=browser.page_source
    final=BeautifulSoup(html_final,'html.parser')
    pattern=re.compile(r'window.res_id\s*\=\s*(\d+);')
    try:
        key=(final.find("script",text=pattern).text)
    except AttributeError:
        return 
    #key=(final.find("script",text=pattern).text)
    res_id=str(int(key[key.find("=")+1:key.find(";")]))
    if res_id in done_rest:
        print("Already Included rest")
        return
    else:
        done_rest+=[res_id]
    menu[res_id]=[]
    #print(done_rest)
    divTag =browser.find_elements_by_class_name("content")
    for tag in divTag:
       tdTags = tag.find_elements_by_class_name("header")
       for tag in tdTags:
            menu[res_id].append(tag.text)
           #print(tag.text)
            menu_format = {"id":res_id,
     "items":[{'name':value} for value in menu[res_id]]}
    #fwrites(menu,res_id)
    try:
        jsonwrites(menu_format)
    except UnboundLocalError:
        return
   
  
    
```
